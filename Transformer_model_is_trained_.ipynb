```python
import numpy as np

# -----------------------------
# 1. Input sentence and tokenization
# -----------------------------
sentence = "What are the symptoms of diabetes?"
tokens = sentence.replace("?", " ?").split()
print("Tokens:", tokens)

# -----------------------------
# 2. Hyperparameters
# -----------------------------
np.random.seed(42)
d_model = 8   # embedding dimension
d_k = 8       # dimension of query/key
n_tokens = len(tokens)

# -----------------------------
# 3. Create toy token embeddings
#    (In a real Transformer, these are learned embeddings)
# -----------------------------
token_embeddings = np.random.randn(n_tokens, d_model)

# -----------------------------
# 4. Positional encoding
# -----------------------------
def positional_encoding(seq_len, d_model):
    PE = np.zeros((seq_len, d_model))
    positions = np.arange(seq_len)[:, np.newaxis]
    div_term = np.exp(np.arange(0, d_model, 2) * (-np.log(10000.0) / d_model))

    PE[:, 0::2] = np.sin(positions * div_term)
    PE[:, 1::2] = np.cos(positions * div_term)
    return PE

pos_encoding = positional_encoding(n_tokens, d_model)

# Final encoder input
X = token_embeddings + pos_encoding

# -----------------------------
# 5. Initialize weight matrices
# -----------------------------
W_Q = np.random.randn(d_model, d_k)
W_K = np.random.randn(d_model, d_k)
W_V = np.random.randn(d_model, d_k)

# -----------------------------
# 6. Compute Q, K, V
# -----------------------------
Q = X @ W_Q
K = X @ W_K
V = X @ W_V

# -----------------------------
# 7. Compute scaled dot-product attention scores
# -----------------------------
scores = (Q @ K.T) / np.sqrt(d_k)

# -----------------------------
# 8. Softmax function
# -----------------------------
def softmax(x):
    exp_x = np.exp(x - np.max(x, axis=-1, keepdims=True))
    return exp_x / np.sum(exp_x, axis=-1, keepdims=True)

attention_weights = softmax(scores)

# -----------------------------
# 9. Compute self-attention output
# -----------------------------
attention_output = attention_weights @ V

# -----------------------------
# 10. Print results
# -----------------------------
np.set_printoptions(precision=3, suppress=True)

print("\nSelf-Attention Score Matrix:")
print(scores)

print("\nAttention Weight Matrix (after softmax):")
print(attention_weights)

print("\nSelf-Attention Output:")
print(attention_output)
```
